# ✅ 1. Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from keras.models import Sequential
from keras.layers import Dense, LSTM
import json
import datetime

# Menampilkan semua kolom
pd.set_option('display.max_columns', None)
```

# ✅ 2. Load dan Pratinjau Dataset JSON

```python
# Load file JSON
with open("dataset_cpo.json", "r") as file:
    data = json.load(file)

df = pd.DataFrame(data['data'])

# Konversi kolom harga dan tanggal
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
df['Penetapan_Harga'] = pd.to_numeric(df['Penetapan_Harga'])

# Sortir berdasarkan tanggal
df = df.sort_values('Tanggal').reset_index(drop=True)

df[['Tanggal', 'Penetapan_Harga']].head()
```

# ✅ 3. Visualisasi Awal Harga

```python
plt.figure(figsize=(14, 5))
plt.plot(df['Tanggal'], df['Penetapan_Harga'], label='Harga CPO')
plt.title("Harga CPO dari Waktu ke Waktu")
plt.xlabel("Tanggal")
plt.ylabel("Harga")
plt.legend()
plt.grid(True)
plt.show()
```

# ✅ 4. Preprocessing untuk LSTM

```python
data = df[['Penetapan_Harga']].values

# Normalisasi
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

# Windowing (menggunakan 60 hari terakhir untuk prediksi 1 hari ke depan)
def create_dataset(dataset, window_size=60):
    X, y = [], []
    for i in range(window_size, len(dataset)):
        X.append(dataset[i-window_size:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

window_size = 90
X, y = create_dataset(scaled_data, window_size)

# Bentuk ke (samples, time_steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))
```

# ✅ 5. Split Data Latih dan Test

```python
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

# ✅ 6. Bangun dan Latih Model LSTM

```python
from keras.layers import Dropout

model = Sequential()
model.add(LSTM(units=128, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=128, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=150, batch_size=32, validation_data=(X_test, y_test))
```

# ✅ 7. Evaluasi Model

```python
y_pred = model.predict(X_test)
y_pred_inv = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"RMSE: {rmse}")
print(f"R2 Score: {r2 * 100:.2f}%")
```

```python
plt.figure(figsize=(14, 5))
plt.plot(df['Tanggal'][-len(y_test):], y_test_inv, label='Actual')
plt.plot(df['Tanggal'][-len(y_test):], y_pred_inv, label='Predicted')
plt.title("Prediksi vs Aktual Harga CPO")
plt.xlabel("Tanggal")
plt.ylabel("Harga")
plt.legend()
plt.grid(True)
plt.show()
```
