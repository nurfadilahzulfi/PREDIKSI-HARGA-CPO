```python
# import library yang dibutukan dan disesuaikan untuk dataset yang bersarang dan dalam bentuk json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
import json
import datetime
import os

# TensorFlow/Keras
try:
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, LSTM, Dropout
    print("‚úÖ Menggunakan tensorflow.keras")
except ImportError:
    try:
        from keras.models import Sequential
        from keras.layers import Dense, LSTM, Dropout
        print("‚úÖ Menggunakan keras standalone")
    except ImportError:
        print("‚ùå Error: Tidak dapat mengimpor Keras/TensorFlow")
        exit(1)

# Pickle & Joblib
try:
    import pickle
    PICKLE_AVAILABLE = True
    print("‚úÖ Pickle tersedia")
except ImportError:
    PICKLE_AVAILABLE = False
    print("‚ö†Ô∏è Pickle tidak tersedia")

try:
    import joblib
    JOBLIB_AVAILABLE = True
    print("‚úÖ Joblib tersedia")
except ImportError:
    JOBLIB_AVAILABLE = False
    print("‚ö†Ô∏è Joblib tidak tersedia")

if not PICKLE_AVAILABLE and not JOBLIB_AVAILABLE:
    print("‚ùå Error: Tidak ada library untuk menyimpan scaler (pickle atau joblib)")
    exit(1)

# Tampilkan semua kolom
pd.set_option('display.max_columns', None)

print("üöÄ Memulai program training dan ekspor model LSTM...")
print("="*60)
```

```python
# ===================== LOAD & EKSTRAK DATASET =====================
try:
    with open("dataset_cpo.json", "r") as file:
        raw_data = json.load(file)

    records = []
    for year in raw_data["data"]["years"]:
        for month in year["months"]:
            for entry in month["detail"]:
                if entry["value"] == 0:
                    continue
                records.append({
                    "Tanggal": entry["tanggal"],
                    "Penetapan_Harga": entry["value"]
                })

    df = pd.DataFrame(records)
    print("‚úÖ Dataset berhasil dimuat dan diekstrak")

    # ‚úÖ Tambahan: tampilkan 10 data pertama
    print("\nüîç Contoh 10 data pertama:")
    print(df.head(10))

except Exception as e:
    print(f"‚ùå Gagal memuat dataset: {e}")
    exit(1)
```

```python
# ===================== KONVERSI DAN SORT =====================
try:
    df['Tanggal'] = pd.to_datetime(df['Tanggal'])
    df['Penetapan_Harga'] = pd.to_numeric(df['Penetapan_Harga'])
    print("‚úÖ Konversi data berhasil")
except Exception as e:
    print(f"‚ùå Error konversi data: {e}")
    exit(1)

df = df.sort_values('Tanggal').reset_index(drop=True)

print(f"üìä Total data: {len(df)} records")
print(f"üìÖ Rentang tanggal: {df['Tanggal'].min()} - {df['Tanggal'].max()}")
print(f"üí∞ Rentang harga: Rp {df['Penetapan_Harga'].min():,.0f} - Rp {df['Penetapan_Harga'].max():,.0f}")

# ===================== VISUALISASI =====================
plt.figure(figsize=(14, 5))
plt.plot(df['Tanggal'], df['Penetapan_Harga'], label='Harga CPO', color='blue', linewidth=1)
plt.title("Harga CPO dari Waktu ke Waktu")
plt.xlabel("Tanggal")
plt.ylabel("Harga (Rp)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# ===================== PERSIAPAN DATA =====================
print("\nüîÑ Memproses data untuk training...")
data = df[['Penetapan_Harga']].values

scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
print("‚úÖ Normalisasi data selesai")

def create_dataset(dataset, window_size=60):
    X, y = [], []
    for i in range(window_size, len(dataset)):
        X.append(dataset[i - window_size:i, 0])
        y.append(dataset[i, 0])
    return np.array(X), np.array(y)

window_size = 90
X, y = create_dataset(scaled_data, window_size)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print(f"üìä Data training: {X_train.shape[0]} samples")
print(f"üìä Data testing : {X_test.shape[0]} samples")

# ===================== BANGUN MODEL =====================
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(X.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(128, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
print("‚úÖ Model berhasil dibuat")

# ===================== TRAINING =====================
print("\nüéØ Training model...")
history = model.fit(X_train, y_train, epochs=150, batch_size=32, validation_data=(X_test, y_test), verbose=1)
print("‚úÖ Training selesai!")

# ===================== EVALUASI =====================
y_pred = model.predict(X_test)
y_pred_inv = scaler.inverse_transform(y_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv))
r2 = r2_score(y_test_inv, y_pred_inv)

print(f"üìà RMSE     : Rp {rmse:,.2f}")
print(f"üìà R¬≤ Score : {r2 * 100:.2f}%")

# ===================== EKSPOR MODEL =====================
print("\nüíæ Menyimpan model dan data...")

try:
    model.save('cpo_lstm_model.h5')
    print("‚úÖ Model disimpan sebagai 'cpo_lstm_model.h5'")
except Exception as e:
    print(f"‚ùå Gagal menyimpan model: {e}")

try:
    if PICKLE_AVAILABLE:
        with open('scaler.pkl', 'wb') as f:
            pickle.dump(scaler, f)
        print("‚úÖ Scaler disimpan (pickle)")
    elif JOBLIB_AVAILABLE:
        joblib.dump(scaler, 'scaler.joblib')
        print("‚úÖ Scaler disimpan (joblib)")
except Exception as e:
    print(f"‚ùå Gagal menyimpan scaler: {e}")

try:
    df.to_csv('processed_cpo_data.csv', index=False)
    print("‚úÖ Dataset diproses disimpan")
except Exception as e:
    print(f"‚ùå Gagal simpan dataset: {e}")

try:
    np.save('last_data_for_prediction.npy', scaled_data[-window_size:].reshape(1, window_size, 1))
    print("‚úÖ Data terakhir disimpan untuk prediksi")
except Exception as e:
    print(f"‚ùå Gagal menyimpan data prediksi: {e}")

try:
    with open('training_history.json', 'w') as f:
        json.dump({
            'loss': list(map(float, history.history['loss'])),
            'val_loss': list(map(float, history.history['val_loss']))
        }, f, indent=4)
    print("‚úÖ History training disimpan")
except Exception as e:
    print(f"‚ùå Gagal menyimpan history: {e}")
```

```python
# ===================== VISUALISASI HASIL =====================
plt.figure(figsize=(15, 6))
test_dates = df['Tanggal'][-len(y_test):].values
plt.plot(test_dates, y_test_inv, label='Actual', color='blue')
plt.plot(test_dates, y_pred_inv, label='Predicted', color='red')
plt.title(f"Prediksi vs Aktual Harga CPO\nRMSE: Rp {rmse:,.0f}, R¬≤: {r2*100:.1f}%")
plt.xlabel("Tanggal")
plt.ylabel("Harga (Rp)")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
